## 2. Data Pre-Processing

```python
data = data_raw
data = data_raw.loc[np.random.choice(data_raw.index, size=2000)]
data.shape
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem.snowball import SnowballStemmer
import re

import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

### 2.1. Cleaning Data

```python
def cleanHtml(sentence):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ', str(sentence))
    return cleantext


def cleanPunc(sentence): #function to clean the word of any punctuation or special characters
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned


def keepAlpha(sentence):
    alpha_sent = ""
    for word in sentence.split():
        alpha_word = re.sub('[^a-z A-Z]+', ' ', word)
        alpha_sent += alpha_word
        alpha_sent += " "
    alpha_sent = alpha_sent.strip()
    return alpha_sent
```

```python
data['comment_text'] = data['comment_text'].str.lower()
data['comment_text'] = data['comment_text'].apply(cleanHtml)
data['comment_text'] = data['comment_text'].apply(cleanPunc)
data['comment_text'] = data['comment_text'].apply(keepAlpha)
data.head()
```

### 2.2. Removing Stop Words

```python
stop_words = set(stopwords.words('english'))
stop_words.update(['zero','one','two','three','four','five','six','seven','eight','nine','ten','may','also','across','among','beside','however','yet','within'])
re_stop_words = re.compile(r"\b(" + "|".join(stop_words) + ")\\W", re.I)
def removeStopWords(sentence):
    global re_stop_words
    return re_stop_words.sub(" ", sentence)

data['comment_text'] = data['comment_text'].apply(removeStopWords)
data.head()
```

### 2.3. Stemming

```python
stemmer = SnowballStemmer("english")
def stemming(sentence):
    stemSentence = ""
    for word in sentence.split():
        stem = stemmer.stem(word)
        stemSentence += stem
        stemSentence += " "
    stemSentence = stemSentence.strip()
    return stemSentence

data['comment_text'] = data['comment_text'].apply(stemming)
data.head()
```

### 2.4. Train-Test Split

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, random_state=42, test_size=0.30, shuffle=True)

print(train.shape)
print(test.shape)
```

### 2.5. TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2')
vectorizer.fit(train_text)
vectorizer.fit(test_text)
```

```python
x_train = vectorizer.transform(train_text)
y_train = train.drop(labels = ['id','comment_text'], axis=1)
# print(x_train)
print(type(x_train))

x_test = vectorizer.transform(test_text)
y_test = test.drop(labels = ['id','comment_text'], axis=1)
```

## JSON file

```python
from pandas import DataFrame
def createModelDF(accuracy,cost, task):
    model_description_df = DataFrame(columns=['Task','Accuracy','Cost(ms)'])
    model_description_df['Task'] = task
    model_description_df['Accuracy'] = accuracy
    model_description_df['Cost(ms)'] = cost
    return model_description_df
```

```python
import json
import os
import pandas as pd
import numpy as np

class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(NpEncoder, self).default(obj)

def createJSONFile(model, accuracy, cost, task):
    path = 'model/' + model + '.json'
    mode = 'w' #'a' if os.path.isfile(path) else 'w'
    
    # retrieve ID
    summary_path = 'model/ModelSummary.csv'
    if os.path.isfile(summary_path):
        df = pd.read_csv(summary_path)
        ID = df['ID'].iloc[-1] + 1
        df.loc[len(df)] = [ID, model]
    else:
        df = DataFrame(columns=['ID','Model'])
        ID = 1
        df['ID'] = [ID]
        df['Model'] = [model]
    df.to_csv(summary_path, index=False)
    data = {}
    data['id'] = ID
    data['dataset'] = 'Toxic Comment'
    data['task'] = {}
    model_description_df = createModelDF(accuracy, cost, task)
    for t in task:
        data['task'][t] = {}
        data['task'][t]['accuracy'] = model_description_df['Accuracy'][model_description_df['Task'] == t].values[0]
        data['task'][t]['cost'] = model_description_df['Cost(ms)'][model_description_df['Task'] == t].values[0]
    with open( path, mode) as outfile:
        json.dump(data, outfile, cls=NpEncoder)
```

## 3. Multi-Label Classification

### 3.1. Multiple Binary Classifications - (One Vs Rest Classifier)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
```

```python
# %%time
import time
NUMBER_OF_ROUND = 10

# Using pipeline for applying logistic regression and one vs rest classifier
LogReg_pipeline = Pipeline([
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),
            ])

for category in categories:
    accuracy = []
    cost = []
    printmd('**Processing {} comments...**'.format(category))
    
    # Training logistic regression model on train data
    start = time.time()
    for i in range(NUMBER_OF_ROUND):
        LogReg_pipeline.fit(x_train, train[category])
    train_time = time.time() - start
    
    # calculating test accuracy
    start = time.time()
    for i in range(NUMBER_OF_ROUND):
        prediction = LogReg_pipeline.predict(x_test)
    inference_time = round(time.time() - start,5)
    cost.append(inference_time/NUMBER_OF_ROUND)
    
    acc = round(accuracy_score(test[category], prediction),3)
    accuracy.append(acc)
    print('Test accuracy is {}'.format(acc))
    print("Train time is %f seconds and inference time is %f seconds" % (train_time, inference_time)) #/test[category].size))
    
    time.sleep(0.5)
    createJSONFile("OvR_"+category, accuracy, cost, [category])
```

### 3.2. Multiple Binary Classifications - (Binary Relevance)

```python
import sys
!{sys.executable} -m pip install scikit-multilearn
```

```python
# %%time

# using binary relevance
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB

# initialize binary relevance multi-label classifier
# with a gaussian naive bayes base classifier
classifier = BinaryRelevance(GaussianNB())

# train
start = time.time()
for i in range(NUMBER_OF_ROUND):
    classifier.fit(x_train, y_train)
train_time = time.time() - start

# predict
start = time.time()
for i in range(NUMBER_OF_ROUND):
    predictions = classifier.predict(x_test)
inference_time = time.time() - start
```

```python
accuracy, cost = [], []
for idx, category in enumerate(categories):
    printmd('**Processing {} comments...**'.format(category))
    acc = round(accuracy_score(test[category], predictions[:,idx].todense()),3)
    accuracy.append(acc)
    cost.append(round(inference_time/NUMBER_OF_ROUND,5))
    print('Test accuracy is {}'.format(acc))
    print("Train time is %f seconds and inference time is %f seconds" % (train_time, inference_time)) #/test[category].size))

createJSONFile("BinaryRelevance", accuracy, cost, categories)
# accuracy
print("Accuracy = ",accuracy_score(y_test,predictions))
print("\n")
```

### 3.3. Classifier Chains

```python
# using classifier chains
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import LogisticRegression
```

```python
# %%time

# initialize classifier chains multi-label classifier
classifier = ClassifierChain(LogisticRegression())

# Training logistic regression model on train data
start = time.time()
for i in range(NUMBER_OF_ROUND):
    classifier.fit(x_train, y_train)
train_time = time.time() - start

# predict
start = time.time()
for i in range(NUMBER_OF_ROUND):
    predictions = classifier.predict(x_test)
inference_time = time.time() - start
```

```python
cost = [inference_time] * len(categories)
print(cost)
```

```python
print("Train time is %f seconds and inference time is %f seconds" % (train_time, inference_time)) #/test[category].size))
cost = [round(inference_time/NUMBER_OF_ROUND,5)] * len(categories)
accuracy = []
for idx, category in enumerate(categories):
    printmd('**Processing {} comments...**'.format(category))
    acc = round(accuracy_score(test[category], predictions[:,idx].todense()),3)
    accuracy.append(acc)
    print('Test accuracy is {}'.format(acc))

createJSONFile("ClassifierChain", accuracy, cost, categories)
# accuracy
print("Accuracy = ",accuracy_score(y_test,predictions))
print("\n")
```

### 3.5. Adapted Algorithm

```python
# http://scikit.ml/api/api/skmultilearn.adapt.html#skmultilearn.adapt.MLkNN

from skmultilearn.adapt import MLkNN
from scipy.sparse import csr_matrix, lil_matrix
```

```python
#%%time

classifier_new = MLkNN(k=10)

# Note that this classifier can throw up errors when handling sparse matrices.

x_train = lil_matrix(x_train).toarray()
y_train = lil_matrix(y_train).toarray()
x_test = lil_matrix(x_test).toarray()

# train
start = time.time()
for i in range(NUMBER_OF_ROUND):
    classifier_new.fit(x_train, y_train)
train_time = time.time() - start

# predict
start = time.time()
for i in range(NUMBER_OF_ROUND):
    predictions_new = classifier_new.predict(x_test)
inference_time = time.time() - start
```

```python
accuracy=[]
for idx, category in enumerate(categories):
    printmd('**Processing {} comments...**'.format(category))
    acc = round(accuracy_score(test[category], predictions_new[:,idx].todense()),3)
    accuracy.append(acc)
    print('Test accuracy is {}'.format(acc))
    print("Train time is %f seconds and inference time is %f seconds" % (train_time, inference_time)) #/test[category].size))
cost = [round(inference_time/NUMBER_OF_ROUND,5)] * len(categories)
createJSONFile("MLkNN", accuracy, cost, categories)

# accuracy
print("Accuracy = ",accuracy_score(y_test,predictions_new))
print("\n")
```
